# Pandas

```python
import pandas as pd
#Has fucntions : read_csv , read_html , read_json ...etc
#Works with internet links:
iris_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
print(iris_data) #1st row works as a header (title)
print("Type of data : " , type(iris_data))
```

```python
#To make a copy of iris_data:
df = iris_data.copy()
#To take a brife idea about data:
print(df.head(),'\n') #can pass how many rows to see

#Now in order to change the header (column type) of the data :
df.columns = ['Sepal_Length' , 'Sepal_Width' , 'Petal_Length' , 'Petal_Width' , 'Flower_type']

#For shape: use df.shape
print("Shape : " , df.shape,"\n")
#For printing type of data in each row:
print("Types :\n",df.dtypes)
```

```python
#To describe the data:
df.describe() #(% : percentile)
```

```python
#Access a column
print(df.Sepal_Length) 
#Or: 
print(df["Sepal_Length"])
```

```python
#To find is null true or false
df.isnull()
```

```python
#Also we can apply isnull to sum column wise:
df.isnull().sum()
```

```python
#To access slicing in pandas: iloc[]:
df.iloc[1:4 , 2:3]
```

```python
#To get as list of lists: df.values
ctr=1
for i in df.values:
    print(ctr," : ",i)
    ctr += 1
```

```python
#Just like head function there is a tail fucntion:
df.tail(20) #Number of last elements can be passed. Default = 5.
```
